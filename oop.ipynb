```python
def longest_palindrome(s):
    res = ""
    
    for i in range(len(s)):
        
        # odd length
        l = r = i
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if len(res) < r - l + 1:
                res = s[l:r+1]
            l -= 1
            r += 1
        
        # even length
        l, r = i, i + 1
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if len(res) < r - l + 1:
                res = s[l:r+1]
            l -= 1
            r += 1
    
    return res
a="babap"
print(longest_palindrome(a))
```

```python
def second_largest(nums):
    first=second=third=float("-inf")
    for num in nums:
        if num in (first,second,third):
            continue
        if num>first:
            third=second
            second=first
            first=num
        elif third<num and second <num:
            third=second 
            second=num
        elif third<num and second >num:
            third=num
    if third == float("-inf"):
        return "not second value"
    return third
a=[2,2,3,1]
print(second_largest(a))

#space complexity O(1)
#time O(n)
```

```python
class Node:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        self.next=None
        self.prev=None
        
class LRU:
    def __init__(self,capacity):
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        
        self.head.next=self.tail
        self.tail.prev=self.head
    
    def _add_to_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        self.head.next=node
        node.prev=self.head
    
    def _remove(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=prev
    
    def get(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self._remove(node)
        self._add_to_front(node)
        return node.value
    
    def put(self,key,value):
        if key in self.cache:
            node=self.cache[key]
            self.remove(node)
        node=Node(key,value)
        self.cache[key]=node
        self._add_to_front(node)
        if len(self.cache)>self.capacity:
            last=self.tail.prev
            self._remove(last)
            del self.cache[last.key]
    
    def display(self):
        if len(self.cache)==0:
            return "not data in lru"
        curr=self.head.next
        while curr != self.tail:
            print(f"{curr.key}:{curr.value}",end=" ")
            curr=curr.next
        print("END")
lru=LRU(3)
lru.put(1,11)
lru.put(2,22)
lru.put(3,33)
lru.get(1)
lru.put(4,44)
lru.display()
```

```python
def top_k(nums,k):
    import heapq
    s={}
    for right in range(len(nums)):
        s[nums[right]]=s.get(nums[right],0)+1
        
    heap=[]
    for i,num in s.items():
        heapq.heappush(heap,(num,i))
        
        if len(heap)>k:
            heapq.heappop(heap)
    return heap
a=[2,3,4,4,4,1,1,1,1,4]
print(top_k(a,2))        
    
```

```python
def temperature(nums):
    rev=[0]*len(nums)
    stack=[]
    for i in range(len(nums)-1,-1,-1):
        while stack and nums[stack[-1]]<=nums[i]:
            stack.pop()
        if  stack:
            rev[i]=stack[-1]-i
        stack.append(i)
    return rev
a=[73,74,75,71,69,72,76,73]
print(temperature(a))
```

```python
def hhhh(nums,k):
    import heapq
    heap={}
    for i in nums:
        if  i in heap:
            heap[i]+=1
        else:
            heap[i]=1
    hash=[]
    for i,num in heap.items():
        heapq.heappush(hash,(num,i))
        
        if len(hash)>k:
            heapq.heappop(hash)
    return [i for _,i in hash]
a=[4,4,4,3,3,3,1,2]
print(hhhh(a,2))
```

```python
def top_k(nums,k):
    hash={}
    for i in nums:
        if i in hash:
            hash[i]+=1
        else:
            hash[i]=1
    import heapq
    heap=[]
    for i,num in hash.items():
        heapq.heappush(heap,(num,i))
        
        while len(heap)>k:
            heapq.heappop(heap)
    return [x for _,x in heap]
a=[1,2,1,1,1,2,3,3,3,3,4,4,4]
print(top_k(a,2))
```

```python
def minMirrorPairDistance( nums):
    def reverse_num(x):
        return int(str(x)[::-1])
    mini_dist=float("inf")
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[j]==reverse_num(nums[i]):
                mini_dist=min(mini_dist,abs(j-i))
                if mini_dist==1:
                    return 1
    if float("inf")==mini_dist:
        return -1
    return mini_dist
a=[12,21,33,45,54]
print(minMirrorPairDistance(a))
```

```python
def two_pointer(nums,target):
    left=0
    right=len(nums)-1
    while left<=right:
        curr_sum=nums[left]+nums[right]
        if curr_sum==target:
            return (left,right)
        elif curr_sum<target:
            left+=1
        else:
            right-=1
    return "target is not found"
a=[1,2,3,4,5,6,7,8,9] 
print(two_pointer(a,17))
```

```python
def removeAnagram(words):
    j=0
    while j<len(words)-1:
        if sorted(words[j])==sorted(words[j+1]):
            words.pop(j+1)
        else:
            j+=1
    return words
a=["abba","baba","baba","cd","cd"]
print(removeAnagram(a))
```

```python
def moveszero(nums):
    k=0
    for i in range(len(nums)):
        if nums[i]!=0:
            nums[k]=nums[i]
            k+=1
    while k<len(nums):
        nums[k]=0
        k+=1
    return nums
a=[3,0,2,0,1,0,8]
print(moveszero(a))
```

```python
def isPalindrome(x):
        if x>0:
            sign=1
        else:
            sign=-1
        y=x
        x=abs(x)
        result=0
        while x > 0:
            digit=x%10
            result=result*10+digit
            x=x//10
        return result*sign==y
print(isPalindrome(-122))
```

```python
def detect_cycle(self):
    slow=self.head
    fast=self.head
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
        
        if fast.next==slow:
            fast.next=None
    return self.head
```
