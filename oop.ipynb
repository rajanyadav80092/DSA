```python
#ENCAPSULATION 
class Bank:
     
    bank_name="bank of baroda"
    all_acounts=[]
    
    def __init__(self,balance,account,name,age):
        self.__balance=balance
        self.account=account
        self.name=name
        self.age=age
        Bank.all_acounts.append(self)
    
    def deposit(self,amount):
        self.__balance+=amount
        print(f"your total balance is {self.__balance}")
    
    def withdrawn(self,amount):
        if self.__balance<=amount:
            return print("please check your balance")
        self.__balance-=amount
        print(f"your total balance is {self.__balance}")
        
    @property   
    def show_details(self):
        return print("name:",self.name,"balance:" ,self.__balance,"age:",self.age,"account :",self.account)
    
    @staticmethod
    def desclaimer():
        print("your try to account then your minimum balance is required to 1000 and your age is minimum 12 years old")
    
    @classmethod   
    def all_account(cls):
        print(f"\nAll Accounts in {cls.bank_name}:")
        for user in cls.all_acounts:
            print(user.show_details)
        
    @classmethod
    def bankname(cls,new_name):
        cls.bank_name=new_name
        print(new_name)
        
    @property
    def get_balance(self):
        return print(f"your total balance is {self.__balance}")
b=Bank(1000,12345,"pawan",21)
b1=Bank(1000,12335,"vikash",12)
b1.show_details
b.show_details
b.withdrawn(150)
b.deposit(1500)
b.withdrawn(1500)
b.desclaimer()
Bank.bankname("hdfc")
b.get_balance
Bank.all_account()
```

```python
def longest_sum(nums,k):
    left=0
    leng=0
    sum=0
    for right in range(len(nums)):
        sum+=nums[right]
        while sum==k:
            sum-=nums[left]
            left+=1
        leng=max(leng,right-left+1)
    return leng
a=[1,-1,5,-2,3]
print(longest_sum(a,3))
```

```python
def reverse_list(s):
    words=s.strip().split()
    left,right=0,len(words)-1
    while left<right:
        words[left],words[right]=words[right],words[left]
        left+=1
        right-=1
    return " ".join(words)
a="this is apple"
print(reverse_list(a))
```

```python
def longest_palindrome(s):
    res = ""
    
    for i in range(len(s)):
        
        # odd length
        l = r = i
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if len(res) < r - l + 1:
                res = s[l:r+1]
            l -= 1
            r += 1
        
        # even length
        l, r = i, i + 1
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if len(res) < r - l + 1:
                res = s[l:r+1]
            l -= 1
            r += 1
    
    return res
a="babap"
print(longest_palindrome(a))
```

```python
def second_largest(nums):
    first=second=third=float("-inf")
    for num in nums:
        if num in (first,second,third):
            continue
        if num>first:
            third=second
            second=first
            first=num
        elif third<num and second <num:
            third=second 
            second=num
        elif third<num and second >num:
            third=num
    if third == float("-inf"):
        return "not second value"
    return third
a=[2,2,3,1]
print(second_largest(a))

#space complexity O(1)
#time O(n)
```

```python
class Node:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        self.next=None
        self.prev=None
        
class LRU:
    def __init__(self,capacity):
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        
        self.head.next=self.tail
        self.tail.prev=self.head
    
    def _add_to_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        self.head.next=node
        node.prev=self.head
    
    def _remove(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=prev
    
    def get(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self._remove(node)
        self._add_to_front(node)
        return node.value
    
    def put(self,key,value):
        if key in self.cache:
            node=self.cache[key]
            self.remove(node)
        node=Node(key,value)
        self.cache[key]=node
        self._add_to_front(node)
        if len(self.cache)>self.capacity:
            last=self.tail.prev
            self._remove(last)
            del self.cache[last.key]
    
    def display(self):
        if len(self.cache)==0:
            return "not data in lru"
        curr=self.head.next
        while curr != self.tail:
            print(f"{curr.key}:{curr.value}",end=" ")
            curr=curr.next
        print("END")
lru=LRU(3)
lru.put(1,11)
lru.put(2,22)
lru.put(3,33)
lru.get(1)
lru.put(4,44)
lru.display()
```

```python
def slinding_window(nums,k):
    sum=0
    count=0
    left=0
    for right in range(len(nums)):
        sum+=nums[right]
        
        while sum>k:
            sum-=nums[left]
            left+=1
        if sum==k:
            count+=1
    return count
a=[1,1,1]
print(slinding_window(a,2))
```

```python
def negative_move(nums):
    left,right=0,len(nums)-1
    while left<=right:
        if nums[left]>0:
            left+=1
        elif nums[right]<0:
            right-=1
        else:
            nums[left],nums[right]=nums[right],nums[left]
    return nums
a=[-3,5,7,-9,2,4,-8,1]
print(negative_move(a))
```
