```python
#ENCAPSULATION 
class Bank:
     
    bank_name="bank of baroda"
    all_acounts=[]
    
    def __init__(self,balance,account,name,age):
        self.__balance=balance
        self.account=account
        self.name=name
        self.age=age
        Bank.all_acounts.append(self)
    
    def deposit(self,amount):
        self.__balance+=amount
        print(f"your total balance is {self.__balance}")
    
    def withdrawn(self,amount):
        if self.__balance<=amount:
            return print("please check your balance")
        self.__balance-=amount
        print(f"your total balance is {self.__balance}")
        
    @property   
    def show_details(self):
        return print("name:",self.name,"balance:" ,self.__balance,"age:",self.age,"account :",self.account)
    
    @staticmethod
    def desclaimer():
        print("your try to account then your minimum balance is required to 1000 and your age is minimum 12 years old")
    
    @classmethod   
    def all_account(cls):
        print(f"\nAll Accounts in {cls.bank_name}:")
        for user in cls.all_acounts:
            print(user.show_details)
        
    @classmethod
    def bankname(cls,new_name):
        cls.bank_name=new_name
        print(new_name)
        
    @property
    def get_balance(self):
        return print(f"your total balance is {self.__balance}")
b=Bank(1000,12345,"pawan",21)
b1=Bank(1000,12335,"vikash",12)
b1.show_details
b.show_details
b.withdrawn(150)
b.deposit(1500)
b.withdrawn(1500)
b.desclaimer()
Bank.bankname("hdfc")
b.get_balance
Bank.all_account()
```

```python
# MULTILEVEL INHERITANCE A TO B TO C
class grandfather:
    def property(self):
        print("grand father : farmer land in village")
class father(grandfather):
    def Car(self):
        print("father : have a car")
class uncle(grandfather):
    def scooter(self):
        print("father have a scoter")
class cousin(uncle):
    def phone(self):
        print("cousin have  a smart phone ")
class child(father):
    def laptop(self):
        print("son: have a laptop")
obj2=cousin()
obj2.property()
obj2.scooter()
obje=child()
obje.property()
obje.Car()
```

```python
#SINGLE INHERITANCE
class parent:
    @property
    def father(self):
        print("this is father")
    
class child(parent):
    @property
    def son(self):
        print("this is son")

c=child()
c.father
c.son
```

```python
#MULTIPLE INHERITANCE 
class father():
    def doctor(self):
        print("father is doctor")
class mother():
    def engineer(self):
        print("mother is engineer")
class child(mother,father):
    def student(self):
        print("child is student")
c=child()
c.doctor()
c.engineer()
c.student()
```

```python
def longest_common_string(nums):
    for i in range(len(nums[0])):
        ch=nums[0][i]
    
        for j in nums[1:]:
            if len(j)==i or  j[i]!=ch:
                return j[0][:i]
    return nums[0]
arr=["pple","airoplane","artificial"]
print(longest_common_string(arr))
```

```python
def mini(nums):
    min=nums[0]
    for i in nums:
        if i<min:
            min=i
    return min
a=[4,5,6,7,8,9,0]
print(mini(a))
```

```python
def merge_sort(num):
    if len(num)<=1:
        return num
    mid=len(num)//2
    left=merge_sort(num[:mid])
    right=merge_sort(num[mid:])
    return merge(left,right)
def merge(left,right):
    result=[]
    i=j=0
    while i<len(left) and j<len(right):
        if left[i]<right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
a=[9,4,3,4,1,2,6,7,8]
print(merge_sort(a))
```
