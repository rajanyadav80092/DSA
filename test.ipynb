```python
def prefix_suffix(nums):
    n=len(nums)
    ans=[1]*n
    
    prefix=1
    for i in range(n):
        ans[i]=prefix
        prefix*=nums[i]
    
    suffix=1
    for i in range(n-1,-1,-1):
        ans[i]*=suffix
        suffix*=nums[i]
    return ans
a=[1,2,3,4]
print(prefix_suffix(a))
```

```python
import heapq
def topk(nums,k):
    from collections import Counter
    num=Counter(nums)
    arr=[]
    for i,n in num.items():
        heapq.heappush(arr,(i,n))
        
    for i in range(len(arr)-k):
        heapq.heappop(arr)
    return arr
a=[3,9,8,5,1,6]
print(topk(a,2))
        
```

```python
class Node:
    def __init__(self,key,data):
        self.key=key
        self.data=data
        self.next=None
        self.prev=None

class LRU:
    def __init__(self,capacity):
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        self.head.next=self.tail
        self.tail.prev=self.head
    
    def add_to_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        node.prev=self.head
        self.head.next=node
    
    def remove(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=prev
    
    def get_key(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self.remove(node)
        self.add_to_front(node)
        return node.data
    
    def put(self,key,data):
        if key in self.cache:
            node=self.cache[key]
            node.data=data
            self.remove(node)
            self.add_to_front(node)
            return 
        node=Node(key,data)
        self.cache[key]=node
        self.add_to_front(node)
        if len(self.cache)>self.capacity:
            last=self.tail.prev
            self.remove(last)
            del self.cache[last.key]
    
    def display(self):
        curr=self.head.next
        while curr!=self.tail:
            print(f"{curr.key}:{curr.data}",end=" ")
            curr=curr.next
        print("END")
        
        
lru=LRU(3)
lru.put(1,11)
lru.put(2,22)
lru.put(3,33)

lru.get_key(1)
lru.put(2,222)
lru.put(4,44)

lru.display()
            
            
```

```python
class Node:
    def __init__(self,key,data):
        self.key=key
        self.data=data
        self.next=None
        self.prev=None
        
class LRU:
    def __init__(self,capacity):
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        self.head.next=self.tail
        self.tail.prev=self.head
    
    def add_to_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        node.prev=self.head
        self.head.next=node
        
    
    def remove_node(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=prev
    
    def get_key(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self.remove_node(node)
        self.add_to_front(node)
        return node.data
    
    def put(self,key,data):
        if key in self.cache:
            node=self.cache[key]
            self.remove_node(node)
            node.data=data
            self.add_to_front(node)
            return 
        node=Node(key,data)
        self.cache[key]=node
        self.add_to_front(node)
        if self.capacity<len(self.cache):
            last=self.tail.prev
            self.remove_node(last)
            del self.cache[last.key]
    
    def display(self):
        curr=self.head.next
        while curr!=self.tail:
            print(f"{curr.key}:{curr.data}",end=" ")
            curr=curr.next
        print("END")
lru=LRU(3)
lru.put(1,11)
lru.put(2,22)

lru.put(3,33)
print(lru.get_key(1))
lru.put(4,44)

lru.display()
```

```python
class Solution:

    def first_pos(self, nums, k):
        left, right = 0, len(nums)-1
        ans = -1

        while left <= right:
            mid = (left + right)//2

            if nums[mid] == k:
                ans = mid
                right = mid - 1

            elif nums[mid] < k:
                left = mid + 1

            else:
                right = mid - 1

        return ans

    def last_pos(self, nums, k):
        left, right = 0, len(nums)-1
        ans = -1

        while left <= right:
            mid = (left + right)//2

            if nums[mid] == k:
                ans = mid
                left = mid + 1

            elif nums[mid] < k:
                left = mid + 1

            else:
                right = mid - 1

        return ans

    def countFreq(self, nums, k):

        first = self.first_pos(nums, k)

        if first == -1:
            return 0

        last = self.last_pos(nums, k)

        return last - first + 1
a=[1,2,2,2,2,3,4]
print(countFreq(a,2))
```

```python
def reverseVowels(s):
        vowel=["a","e","i","o","u"]
        left,right=0,len(s)-1
        while left<=right:
            if s[left] in vowel and s[right] in vowel and s[left]!=s[right]:
                s[left],s[right]=s[right],s[left]
                left+=1
                right-=1
            elif s[left] in vowel and s[right] not in vowel:
                right-=1
            elif s[right]  in vowel and s[left] in vowel:
                left+=1
        return s
s = "IceCreAm"
print(reverseVowels(s))
```

```python
def productExceptSelf(nums):
    n = len(nums)
    ans = [1] * n

    prefix = 1
    for i in range(n):
        ans[i] = prefix
        prefix *= nums[i]

    suffix = 1
    for i in range(n - 1, -1, -1):
        ans[i] *= suffix
        suffix *= nums[i]

    return ans
a=[1,2,3,4]
print(productExceptSelf(a))
```

```python
from collections import deque
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
class BST:
    def __init__(self):
        self.root=None
    
    def inserted(self,data):
        new_root=Node(data)
        if self.root is None:
            self.root=new_root
            return 
        temp=self.root
        while True:
            if temp.data>data:
                if temp.left is None:
                    temp.left=new_root
                    return
                temp=temp.left
            elif temp.data<=data:
                if temp.right is None:
                    temp.right=new_root
                    return 
                temp=temp.right
                
    def levelorder(self):
        if self.root is None:
            return "bst is empty"
        res=[]
        queue=deque([self.root])
        while queue:
            l=[]
            n=len(queue)
            for _ in range(n):
                node=queue.popleft()
                l.append(node.data)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            res.append(l)
        return res

bst=BST()
a=[7,4,9,5,8,3,2,12,34,7]
for i in a:
    bst.inserted(i)
print(bst.levelorder())
    
                    
```

```python
class Node:
    def __init__(self,key,data):
        self.key=key
        self.next=None
        self.data=data
        self.prev=None

class LRU:
    def __init__ (self,capacity):
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        
        self.head.next=self.tail
        self.tail.prev=self.head
    
    
    def add_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        self.head.next=node
        node.prev=self.head
    
    def remove(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=prev
    
    def get(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self.remove(node)
        self.add_front(node)
        return  node.data
    
    def put(self,key,data):
        if key in self.cache:
            node=self.cache[key]
            node.data=data
            self.remove(node)
            self.add_front(node)
            return
        node=Node(key,data)
        self.cache[key]=node
        self.add_front(node)
        if self.capacity<len(self.cache):
            last=self.tail.prev
            self.remove(last)
            del self.cache[last.key]
    
    def display(self):
        if len(self.cache)==0:
            return "linked list is empty"
        curr=self.head.next
        while curr != self.tail:
            print(f"{curr.key}:{curr.data}",end=" ")
            curr=curr.next
        print("END")

lru=LRU(3)
lru.put(1,22)
lru.put(2,22)
lru.put(3,33)
lru.put(1,11)
lru.put(4,44)
lru.display()
            
            
        
```

```python
def foursum(nums,target):
    nums.sort()
    arr=[]
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            left=j+1
            right=len(nums)-1
            while left<right:
                ss=target-(nums[i]+nums[j])
                if nums[left]+nums[right]==ss:
                    arr.append([nums[i],nums[j],nums[left],nums[right]])
                    left+=1
                    right-=1
                elif nums[left]+nums[right]>ss:
                    right-=1
                else:
                    left+=1
    return arr
a=[1,0,-1,0,-2,2]
print(foursum(a,0))
```

```python
def thirdlargest(nums):
    first=float("-inf")
    second=float("-inf")
    third=float("-inf")
    for i in nums:
        if i>first:
            third=second
            second=first
            first=i
        elif i>=second and i<first:
            third=second
            second=i
        else:
            third=i
    return -1 if third==float("-inf") else third
a=[-1,-1,-1,6,9,5,6]
print(thirdlargest(a))
```

```python
import heapq
def large(nums,k):
    heap=[]
    for i in nums:
        if i in heap:
            continue
        heapq.heappush(heap,i)
        
        if len(heap)>k:
            heapq.heappop(heap)
    return heap[0]
a=[9,6,6,5]
print(large(a,3))
```

```python
def squareroot(x):
    if x == 0:
        return 0

    guess = x

    while abs(guess * guess - x) > 1e-6:
        guess = (guess + x / guess) / 2
        

    return guess

print(squareroot(16))
```
