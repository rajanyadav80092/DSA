```python
def twopair(nums,target):
    hash={}
    for i,num in enumerate(nums):
        s=target-num
        if s in hash:
            return [hash[s],i]
        hash[num]=i
    return "not found"
a=[-1,0,2,1,4]
print(twopair(a,0))
```

```python
def two(nums,target):
    hash={}
    for i in range(len(nums)):
        n=target-nums[i]
        if n in hash:
            return [hash[n],i]
        else:
            hash[nums[i]]=i
    return "target not found"
a=[-1,0,2,1,4]
print(two(a,0))
```
