```python
#binary search for sorted array but rotate between index to left to right and search index number best type binary search rotate array fast
#time complexity is o(logn)and linear search time complexity is o(n)
def search_rotated_binary(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        # Left half sorted hai
        if arr[left] <= arr[mid]:
            if arr[left] <= target < arr[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # Right half sorted hai
        else:
            if arr[mid] < target <= arr[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1

# Example
arr = [4, 5, 6, 7, 1, 2, 3]
target = 2
print(search_rotated_binary(arr, target))  # Output: 2
```

```python
def is_one_edit_distance(s, t):
    if abs(len(s) - len(t)) > 1:
        return False

    i = j = 0
    diff = 0
    while i < len(s) and j < len(t):
        if s[i] != t[j]:
            if diff == 1:
                return False
            diff += 1
            if len(s) > len(t):
                i += 1
                continue
            elif len(s) < len(t):
                j += 1
                continue
        i += 1
        j += 1

    return diff == 1 or (diff == 0 and abs(len(s) - len(t)) == 1)
words = ["apple", "banana", "orange", "pink"]
target = "banaba"

for word in words:
    if is_one_edit_distance(word, target):
        print(f"{target} differs by one letter from {word}")
```

```python
def infix_to_postfix(expr):
    precedence = {'+':1, '-':1, '*':2, '/':2, '^':3}
    stack = []
    output = ""

    for ch in expr:
        if ch.isalnum():  # A, B, 1, 2 etc.
            output += ch
        elif ch == '(':
            stack.append(ch)
        elif ch == ')':
            while stack and stack[-1] != '(':
                output += stack.pop()
            stack.pop()  # remove '('
        else:  # operator
            while stack and stack[-1] != '(' and precedence.get(ch, 0) <= precedence.get(stack[-1], 0):
                output += stack.pop()
            stack.append(ch)

    while stack:
        output += stack.pop()

    return output

print(infix_to_postfix("(A+B)*C"))
```

```python
def binary_search(num,target,low,high):
    if low>=high:
        return -1
    mid=(low+high)//2
    
    if num[mid]==target:
        return mid
    
    elif num[mid]>=target:
        return binary_search(num,target,low,mid-1)
    else:
        return binary_search(num,target,mid+1,high)
a=[1,2,3,4,5,6,7,8,9]
print(binary_search(a,9,0,len(a)))
        
```

```python
def postfix(num):
    stack=[]
    for i in str(num):
        if i.isdigit():
            stack.append(int(i))
        else:
            b=stack.pop()
            a=stack.pop()
            if i=="+":
                stack.append(a+b)
            elif i == "-":
                stack.append(a-b)
            elif i == "*":
                stack.append(a*b)
            elif i == "/":
                stack.append(a/b)
    return stack[0]
print(postfix("234*+9-"))
```

```python
def is_balanced(s):
    stack = []
    # Mapping closing â†’ opening
    pairs = {')': '(', '}': '{', ']': '['}

    for ch in s:
        if ch in '({[':
            stack.append(ch)
        elif ch in ')}]':
            # agar stack empty hai ya top match nahi karta
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()  # match mil gaya to remove karo
    
    # agar sab pair match ho gaye to stack empty hoga
    return not stack


# Test cases
print(is_balanced("{[()]}"))   # âœ… True
print(is_balanced("{[(])}"))   # âŒ False
print(is_balanced("((()))"))   # âœ… True
print(is_balanced("{[}"))      # âŒ False
```

```python
import heapq
from collections import Counter

def top_k_trending_searches(searches, k):
    # 1ï¸âƒ£ Frequency count nikal lo
    freq = {}
    for i in searches:
        if i in freq:
            freq[i]+=1
        else:
            freq[i]=1
            
    
    # 2ï¸âƒ£ Heap banao (min-heap) with (-count, word)
    # negative count use kar rahe hain taaki max-heap jaisa behave kare
    heap = [(-count, word) for word, count in freq.items()]
    heapq.heapify(heap)
    
    # 3ï¸âƒ£ Top k elements nikaal lo
    top_k = []
    for _ in range(k):
        count, word = heapq.heappop(heap)
        top_k.append((word, -count))
    
    return top_k

# Example data
searches = ["ai", "python", "ai", "dsa", "python", "ai", "ml", "ml","help", "python", "heap","ai"]

print(top_k_trending_searches(searches, 5))
```

```python
import heapq
from collections import Counter

def top_k_words(text, k=10):
    # 1ï¸âƒ£ Convert text to lowercase aur split by space
    words = text.lower().split()

    # 2ï¸âƒ£ Remove punctuation marks
    clean_words = [word.strip(".,!?()[]{}\"'") for word in words]

    # 3ï¸âƒ£ Count frequency using Counter
    freq = Counter(clean_words)

    # 4ï¸âƒ£ Get top K frequent words using heapq.nlargest
    top_k = heapq.nlargest(k, freq.items(), key=lambda x: x[1])

    # 5ï¸âƒ£ Print result
    print(f"Top {k} most frequent words:")
    for word, count in top_k:
        print(f"{word}: {count}")

# Example paragraph
text = """
Python is powerful and Python is fun.
Python can build AI, web apps, and automate boring stuff.
AI and data science use Python a lot!
"""

top_k_words(text, 5)
```

```python
def postfix(num):
    stack=[]
    for i in num:
        if i.isdigit():
            stack.append(int(i))
        else:
            b=stack.pop()
            a=stack.pop()
            if i=="+":
                stack.append(a+b)
            elif i =="-":
                stack.append(a-b)
            elif i =="*":
                stack.append(a*b)
            elif i == "/":
                stack.append(a/b)
    return stack[0]
print(postfix("249*+9-"))
        
```

```python
def binary_search(num,target,low,high):
    if low>high:
        return -1
    mid=(low+high)//2
    if num[mid]==target:
        return mid
    elif num[mid]>=target:
        return binary_search(num,target,low,mid-1)
    else:
        return binary_search(num,target,mid+1,high)
a=[1,2,3,4,5,6,7,8]
print(binary_search(a,1,0,len(a)))
    
```

```python
def merge_sort(num):
    if len(num)<=1:
        return num
    mid=len(num)//2
    left=merge_sort(num[:mid])
    right=merge_sort(num[mid:])
    return merge(left,right)
def merge(left,right):
    results=[]
    i=j=0
    while len(left)>i and len(right)>j:
        if left[i]<right[j]:
            results.append(left[i])
            i+=1
        else:
            results.append(right[j])
            j+=1
    results.extend(left[i:])
    results.extend(right[j:])
    return results
a=[8,9,6,3,2,1,4,5,6,7]
print(merge_sort(a))
        
```

```python
def insertion_sort(num):
    l=len(num)
    for i in range(1,l):
        key=num[i]
        j=i-1
        while j>=0 and num[j]>=key:
            num[j+1]=num[j]
            j-=1
        num[j+1]=key
    return num
a=[8,7,6,5,4,3,2,9,1]
print(insertion_sort(a))
```

```python
def bubble_sort(num):
    l=len(num)
    for i in range(l):
        for j in range(l-i-1):
            if num[j]>num[j+1]:
                num[j+1],num[j]=num[j],num[j+1]
    return num
a=[6,7,8,9,4,3,2,1,5]
print(bubble_sort(a))
```

```python
class Stack:
    def __init__(self):
        self.stack=[]
        self.minstack=[]
        
    def push(self,data):
        self.stack.append(data)
        if not self.minstack or data<=self.minstack[-1]:
            self.minstack.append(data)
        
    def pop(self):
        if self.stack:
            item=self.stack.pop()
            if item==self.minstack[-1]:
                self.minstack.pop()
            return item
    def show(self):
        return self.stack
    def minimum(self):
        return self.minstack[-1]
s=Stack()
s.push(10)
s.push(11)
s.push(12)
s.push(13)
s.push(14)
s.push(15)
s.push(1)
print(s.minimum())
print(s.show())
print(s.pop())
print(s.minimum())
```

```python
def post(num):
    stack=[]
    for i in num:
        if i.isdigit():
            stack.append(int(i))
        else:
            b=stack.pop()
            a=stack.pop()
            if i=="+":
                stack.append(a+b)
            elif i=="-":
                stack.append(a-b)
            elif i =="*":
                stack.append(a*b)
            elif i == "/":
                stack.append(a/b)
    return stack[0]
print(post("231*-9+"))
    
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def josephus(n, k):
    head = Node(1)
    prev = head

    # build circular list
    for i in range(2, n + 1):
        prev.next = Node(i)
        prev = prev.next

    prev.next = head  # circular

    curr = head
    prev = None

    while curr.next != curr:
        # move k-1 steps
        for _ in range(k - 1):
            prev = curr
            curr = curr.next

        # remove kth node
        prev.next = curr.next
        curr = curr.next

    return curr.data


print(josephus(7, 3))
```

```python
def sliding_window(nums,target):
    left=0
    sum=0
    
    for right in range(len(nums)):
        sum+=nums[right]
        
        while sum>target:
            sum-=nums[left]
            left+=1
        if target==sum:
            return (left,right)
    return "target is not found"
a=[3,4,5,1,2,6,8,9]
print(sliding_window(a,23))
```
