```python
def binary_rotated_search(nums,target):
    left,right=0,len(nums)
    
    while left<=right:
        mid=left+(right-left)//2
        if nums[mid]==target:
            return mid
        elif nums[left]<=nums[mid]:
            if nums[left]<=target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left=mid+1
            else:
                right=mid-1
    return -1
a=[5,6,7,8,9,1,2,3,4]
print(binary_rotated_search(a,2))
                
```

```python
from functools import cmp_to_key

def largestNumber(nums):
    # Convert all numbers to string
    nums = list(map(str, nums))
    
    # Custom comparator
    def compare(a, b):
        if a + b > b + a:
            return -1
        else:
            return 1

    # Sort using custom comparator
    nums.sort(key=cmp_to_key(compare))

    # Edge case: if the largest number is "0", return "0"
    if nums[0] == "0":
        return "0"

    # Join and return
    return "".join(nums)
a=[2,4,5,68,7,0]
print(largestNumber(a))
```

```python
def longest_word(s):
    max_word = ""
    curr = ""

    for ch in s + " ":
        if ch != " ":
            curr += ch
        else:
            if len(curr) > len(max_word):
                max_word = curr
            curr = ""

    return max_word
a="i love javascript programming"
print(longest_word(a))
```

```python
def asteroid(nums):
    heap=[]
    for i in nums:
        while heap and i<0 and heap[-1]>i:
            if heap[-1]<abs(i):
                heap.pop()
                continue
            elif heap[-1]==abs(i):
                heap.pop()
            break
        else:
            heap.append(i)
    return heap
a=[5,-5]
print(asteroid(a))
```

```python
def list_min(nums,k):
    import heapq
    from collections import Counter
    map=Counter(nums)
    heap=[]
    for i,num in map.items():
        heapq.heappush(heap,(num,i))
        
        if len(heap)>k:
            heapq.heappop(heap)
            
    
    return heap
a=[3,3,3,3,4,4,4,1,1,1,1,1]
print(list_min(a,2))
```

```python
def longest_subarray(nums):
    if not nums:
        return " "
    for i in range(len(nums[0])):
        ch=nums[0][i]
        
        for j in nums[1: ]:
            if len(j)==i or j[i]!=ch:
                return nums[0][:i]
    return "not found"
a=["flower","flow","flight"]
print(longest_subarray(a))
                
        
```

```python
def merge_sort(nums):
    if len(nums)<=1:
        return nums
    mid=len(nums)//2
    left=merge_sort(nums[:mid])
    right=merge_sort(nums[mid:])
    return merge(left,right)
def merge(left,right):
    i,j=0,0
    result=[]
    while i<len(left) and j<len(right):
        if left[i]<right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
a=[4,3,1,2,5,8,9,7,6]
print(merge_sort(a))
            
    
```

```python
def bubble_sort(nums):
    l=len(nums)
    for i in range(0,l):
        for j in range(l-i-1):
            if nums[j+1]<nums[j]:
                nums[j+1],nums[j]=nums[j],nums[j+1]
    return nums
a=[9,8,7,6,5,4,1,2,3]
print(bubble_sort(a))
```

```python
def insertion(nums):
    for i in range(1,len(nums)):
        key=nums[i]
        j=i-1
        while j>=0 and key<nums[j]:
            nums[j+1]=nums[j]
            j-=1
        nums[j+1]=key
    return nums
a=[7,6,8,9,4,3,2,1,5]
print(insertion(a))
```
