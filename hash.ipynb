```python
def subarray(arr,k):
    left=0
    freq={}
    count=0
    for right in range(len(arr)):
        freq[arr[right]]=freq.get(arr[right],0)+1
        
        if right-left+1>k:
            freq[arr[left]]-=1
            if freq[arr[left]]==0:
                del freq[arr[left]]
            left+=1
        if right-left+1 == k and len(freq)==k:
            count+=1
    return count
arr=[1,2,3,1,2,3,4]
print(subarray(arr,3))         
```

```python
class Queue:
    def __init__(self):
        self.queue=[]
        self.queue2=[]
    def push(self,data):
        self.queue.append(data)
    
    def pop(self):
        if not self.queue2:
            while self.queue:
                self.queue2.append(self.queue.pop())
        if not self.queue2:
            return "queue is empty"
        return self.queue2.pop()
    
    def get(self):
        if self.queue:
            return self.queue2
        return self.queue
    
    def length(self):
        if self.queue2:
            return len(self.queue2)
        return len(self.queue)
q=Queue()
q.push(1)
q.push(2)
q.push(3)
print(q.pop())
print(q.length())
print(q.get())
print(q.pop())
```

```python
class Stack:
    def __init__(self):
        self.stack=[]
        self.minimum=[]
        
    def push(self,data):
        self.stack.append(data)
        if not self.minimum or data<=self.minimum[-1]:
            self.minimum.append(data)  
           
    def pop(self):
        if self.stack:
            data=self.stack.pop()
            if data==self.minimum[-1]:
                self.minimum.pop()
            return data
        return "stack is empty"
    
    def get(self):
        if self.stack:
            return self.stack
        return "stack is empty"
    
    def get_mini(self):
        if self.minimum:
            return self.minimum
        return "stack is empty"
    
    def minim(self):
        if self.minimum:
            return self.minimum[-1]
        return "stack is empty"
    
    def length(self):
        if self.stack:
            return len(self.stack)
        return "stack is empty"
    
s=Stack()
s.push(3)
s.push(5)
s.push(1)
s.push(4)
s.push(6)
print(s.get_mini())
print(s.get())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.get())

            
```

```python
def prefix(str):
    hash={}
    ss=[]
    for i in str:
        for j in i:
            if j in hash:
                hash[j]+=1
            else:
                hash[j]=1
    for k,num in hash.items():
        if num==3:
            ss.append(k)
    return ss
a=["dog","racecar","car"]
print(prefix(a))
```

```python
def maxsumsubarray(nums,k):
    max_sum=sum(nums[:k])
    window_max=max_sum
    
    for i in range(k,len(nums)):
        max_sum+=nums[i]
        max_sum-=nums[i-k]
        window_max=max(window_max,max_sum)
    return window_max
arr = [2, 1, 5, 1, 3, 2] 
k = 3
print(maxsumsubarray(arr,k))

#fixed side window beacause k is fixed
#time complexity O(n)
#space complexity O(1)
```

```python
def count_vowel(nums,k):
    count=0
    vowel="aeiou"
    
    for i in range(k):
        if nums[i] in vowel:
            count+=1
    max_count=count
    for j in range(k,len(nums)):
        if nums[j] in vowel:
            count+=1
        if nums[j-k] in vowel:
            count-=1
        max_count=max(max_count,count)
    return max_count
arr="abcideiof"
print(count_vowel(arr,3))
            
```

```python
from functools import cmp_to_key

def largestNumber(nums):
    # Convert all numbers to string
    nums = list(map(str, nums))
    
    # Custom comparator
    def compare(a, b):
        if a + b > b + a:
            return -1
        else:
            return 1

    # Sort using custom comparator
    nums.sort(key=cmp_to_key(compare))

    # Edge case: if the largest number is "0", return "0"
    if nums[0] == "0":
        return "0"

    # Join and return
    return "".join(nums)
a=[2,4,5,68,7,0]
print(largestNumber(a))
```

```python
def longest_submit(nums):
    for right in range(len(nums[0])):
        ch=nums[0][right]
        
        for left in nums[1:]:
            if len(left)==len(ch) or ch != left[right]:
                return left[ :right]
    return nums[0]
a=["flower","flight","flow"]
print(longest_submit(a))
            
```
