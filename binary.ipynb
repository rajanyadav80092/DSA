```python
def binary_search(nums,target):
    left,right=0,len(nums)-1
    
    while left <= right:
        mid=(left+right)//2
        if nums[mid]==target:
            return mid
        elif nums[left] <=nums[mid]:
            if nums[left]<=target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left=mid+1
            else:
                right=mid-1
    return -1
a=[5,6,7,8,1,2,3,4]
print(binary_search(a,5))
                
```

```python
def lru(data,k):
    hash={}
    left=0
    max_len=0
    for right in range(len(data)):
        hash[data[right]]=hash.get(data[right],0)+1
        while len(hash)>k:
            hash[data[left]]-=1
            if hash[data[left]]==0:
                del hash[data[left]]
            left+=1
        max_len=max(max_len,right-left+1)
    return max_len
a=[1,2,30,27,1,16,2,3,2,1,4,5,1]
print(lru(a,3))
            
```

```python
def quick(nums):
    if len(nums)<=1:
        return nums
    pivot=nums[-1]
    left=[x for x in nums[:-1] if x<=pivot]
    right=[x for x in nums[:-1] if x>pivot]
    return quick(left)+[pivot]+quick(right)
a=[(9,"j"),(4,"g")]
print(quick(a))
```

```python
def merge_sort(nums):
    if len(nums)<=1:
        return nums
    m=len(nums)//2
    left=merge_sort(nums[:m])
    right=merge_sort(nums[m:])
    return merge(left,right)
def merge(l1,l2):
    result=[]
    i,j=0,0
    while i<len(l1) and j<len(l2):
        if l1[i]<l2[j]:
            result.append(l1[i])
            i+=1
        else:
            result.append(l2[j])
            j+=1
    result.extend(l1[i:])
    result.extend(l2[j:])
    return result
a=[(9,"c",0),(9,"d",1),(5,"a",2)]
print(merge_sort(a))
```

```python
def findMin(nums):
    left, right = 0, len(nums) - 1

    while left<right:
        mid=left+(right-left)//2
        if nums[mid]>nums[right]:
            left=mid+1
        else:
            right=mid
    return nums[left]
a=[5,6,7,8,9,1,2,3,4]
print(findMin(a))
```

```python
from collections import defaultdict
def angram(nums):
    group=defaultdict(list)
    for i in nums:
        key="".join(sorted(i))
        group[key].append(i)
    return group.values()
a=["eye","yee","eat","ate","tea"]
print(angram(a))
```

```python
def rotated_array(nums,target):
    left,right=0,len(nums)-1
    while left<=right:
        mid=left+(right-left)//2
        if nums[left] == nums[mid] == nums[right]:
            left += 1
            right -= 1
        if nums[mid]==target:
            return True
        elif nums[left]<nums[mid]:
            if nums[left]<=target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left=mid+1
            else:
                right=mid-1
    return False
a=[1,1,1,1,2,1]
print(rotated_array(a,2))
```

```python
def mostCommonWord(paragraph,banned) :
    
    banned=set(banned)
    for ch in ".' , ":
        paragraph=paragraph.replace(ch," ")
    paragraph=paragraph.lower().split()
    hash={}
    for i in paragraph:
        i=i.lower()
        if i in hash:
            hash[i]+=1
        else:
            hash[i]=1
    
    maxi=0
    a=""
    for num,j in hash.items():
        if num not in banned  and j>maxi:
            a=num
            maxi=j
    return a
a="Bob hit a ball, the hit BALL flew far after it was hit."
banned=["hit"]
print(mostCommonWord(a,banned))
```

```python
#agar negative number hai aur positive bhi hum fixed size window me maximum khojana hai aur maximum longest length target sum k
def prefixFixed(nums,k):
    prefix=0
    mp={}
    maxlen=0
    for i,num in enumerate(nums):
        prefix+=num
        
        if prefix==k:
            maxlen=i+1
        
        if prefix-k in mp:
            maxlen=max(maxlen,i-mp[prefix-k])
            
        if prefix not in mp:
            mp[prefix]=i
    return maxlen if maxlen!=0 else -1
a=[-6,3,5,7,-1,5,-7,8,9,-3]
print(prefixFixed(a,13))
```

```python
def prefixs(nums,k):    #k is fixed 
    n=len(nums)
    prefix=[0]*(n+1)
    
    for i in range(n):
        prefix[i+1]=prefix[i]+nums[i]
    
    max_sum=float("inf")
    for j in range(k,n+1):
        max_sum=min(prefix[j]-prefix[j-k],max_sum)
    return max_sum
a=[-3,45,-3,5,7,-8,6,9,-1,2]
print(prefixs(a,3))
```

```python
def countSubarray(nums,k):
    s=0
    count=0
    left=0
    for i in range(len(nums)):
        s+=nums[i]
        
        while s>k:
            s-=nums[left]
            left+=1
        
        count+=i-left+1
    return count
a=[1,3,1,4,5,3,5,1,3,1,3,4]
print(countSubarray(a,4))
        
```

```python
from collections import deque
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

class BinarySearch:
    def __init__(self):
        self.root=None
        
    def insert(self,data):
        new_root=Node(data)
        if self.root is None:
            self.root=new_root
            return
        temp=self.root
        while True:
            if temp.data<data:
                if temp.right is None:
                    temp.right=new_root
                    return
                temp=temp.right
            else:
                if temp.left is None:
                    temp.left=new_root
                    return 
                temp=temp.left
    
    def kthsmallest(self,k):
        if self.root is None:
            return "Tree is None"
        rev=[]
        curr=self.root
        while curr or rev:
            while curr:
                rev.append(curr)
                curr=curr.left
                
            curr=rev.pop()
            k-=1
            if k==0:
                return curr.data
            curr=curr.right
        return "data not found k is bigger than root data"
    
    
    def zigzag(self):
        if self.root is None:
            return "tree is empty"
        queue=deque([self.root])
        result=[]
        zig=True
        while queue:
            n=len(queue)
            l=[]
            for _ in range(n):
                node=queue.popleft()
                l.append(node.data)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            
            if not zig:
                l.reverse()
            result.append(l)
            zig= not zig
        return result 
    #time complexity O(n) space O(n) zigzag binary search tree me karu ya binary tree me logic same hi hoga 
                
                
bst=BinarySearch()
a=[10,5,3,6,15]
for i in a:
    bst.insert(i)
print(bst.kthsmallest(21))
print(bst.zigzag())

            
            
        
    
```
