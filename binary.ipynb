```python
def binary_search(nums,target):
    left,right=0,len(nums)-1
    
    while left <= right:
        mid=(left+right)//2
        if nums[mid]==target:
            return mid
        elif nums[left] <=nums[mid]:
            if nums[left]<=target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left=mid+1
            else:
                right=mid-1
    return -1
a=[5,6,7,8,1,2,3,4]
print(binary_search(a,5))
                
```

```python
def lru(data,k):
    hash={}
    left=0
    max_len=0
    for right in range(len(data)):
        hash[data[right]]=hash.get(data[right],0)+1
        while len(hash)>k:
            hash[data[left]]-=1
            if hash[data[left]]==0:
                del hash[data[left]]
            left+=1
        max_len=max(max_len,right-left+1)
    return max_len
a=[1,2,30,27,1,16,2,3,2,1,4,5,1]
print(lru(a,3))
            
```

```python
def quick(nums):
    if len(nums)<=1:
        return nums
    pivot=nums[-1]
    left=[x for x in nums[:-1] if x<=pivot]
    right=[x for x in nums[:-1] if x>pivot]
    return quick(left)+[pivot]+quick(right)
a=[(9,"j"),(4,"g")]
print(quick(a))
```

```python
def merge_sort(nums):
    if len(nums)<=1:
        return nums
    m=len(nums)//2
    left=merge_sort(nums[:m])
    right=merge_sort(nums[m:])
    return merge(left,right)
def merge(l1,l2):
    result=[]
    i,j=0,0
    while i<len(l1) and j<len(l2):
        if l1[i]<l2[j]:
            result.append(l1[i])
            i+=1
        else:
            result.append(l2[j])
            j+=1
    result.extend(l1[i:])
    result.extend(l2[j:])
    return result
a=[(9,"c",0),(9,"d",1),(5,"a",2)]
print(merge_sort(a))
```

```python
def rotated_array(nums,target):
    left,right=0,len(nums)-1
    while left<=right:
        mid=left+(right-left)//2
        if nums[left] == nums[mid] == nums[right]:
            left += 1
            right -= 1
        if nums[mid]==target:
            return True
        elif nums[left]<nums[mid]:
            if nums[left]<=target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left=mid+1
            else:
                right=mid-1
    return False
a=[1,1,1,1,2,1]
print(rotated_array(a,2))
```

```python
class Node:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        
        self.next=None
        self.prev=None
    
class LRU:
    def __init__(self,capacity):
        
            
        self.capacity=capacity
        self.cache={}
        
        self.head=Node(0,0)
        self.tail=Node(0,0)
        
        self.head.next=self.tail
        self.tail.prev=self.head
    
    def add_front(self,node):
        node.next=self.head.next
        self.head.next.prev=node
        node.prev=self.head
        self.head.next=node
    
    def remove_node(self,node):
        prev=node.prev
        nxt=node.next
        prev.next=nxt
        nxt.prev=node
    
    def get_key(self,key):
        if key not in self.cache:
            return -1
        node=self.cache[key]
        self.remove_node(node)
        self.add_front(node)
        return node.value
    
    def put(self,key,value):
        if key in self.cache:
            node=self.cache[key]
            node.value=value
            self.remove_node(node)
            self.add_front(node)
            return 
        node=Node(key,value)
        self.cache[key]=node
        self.add_front(node)
        if len(self.cache)>self.capacity:
            last=self.tail.prev
            self.remove_node(last)
            del self.cache[last.key]
    
    def display(self):
        if not self.cache:
            return "LRU is empty"
        temp=self.head.next
        while temp != self.tail:
            print(f"{temp.key} : {temp.value}",end=" ")
            temp=temp.next
        print("END")
lru=LRU(3)
lru.put(1,"Rajan")
lru.put(2,"Vipin")
lru.put(3,"Akash")
lru.put(1,111)
lru.display()
            
```

```python
def fourSum(nums,target):
    arr=[]
    n=len(nums)
    nums.sort()
    for i in range(n):
        ss=target-nums[i]
        for j in range(i+1,n):
            
            if i > 0 and nums[i] == nums[i-1]:
                continue

            if j > i+1 and nums[j] == nums[j-1]:
                continue
            left=j+1
            right=n-1
            while left<right:
                if nums[j]+nums[left]+nums[right]==ss:
                    arr.append([nums[i],nums[j],nums[left],nums[right]])
                    left+=1
                    right-=1
                    while left<right and nums[left]==nums[left-1]:
                        left+=1
                    while left<right and nums[right]==nums[right+1]:
                        right-=1
                elif (nums[j]+nums[left]+nums[right])<ss:
                    left+=1
                else:
                    right-=1
    return arr
nums=[1,2,3,4,5,6]
target=15
print(fourSum(nums,target))
```

```python
def threesum(nums):
    arr=[]
    nums.sort()
    
    for i in range(len(nums)):
        target=-nums[i]
        left,right=i+1,len(nums)-1
        while left<right:
            if nums[left]+nums[right]==target:
                if ([nums[left],nums[right],nums[i]]) not in arr:
                    arr.append([nums[left],nums[right],nums[i]])
                left+=1
                right-=1
            elif nums[left]+nums[right]<target:
                left+=1
            else:
                right-=1
    return arr
             
arr=[-1,0,1,2,-1,-4]
print(threesum(arr))
        
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Singlylinked:
    def __init__(self):
        self.head=None
    
    def insert_at_begining(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node
        return
    def print_forword(self):
        if self.head is None:
            return "list is empty"
        temp=self.head
        while temp:
            print(temp.data,end=" ")
            temp=temp.next
        print("END")
    
    def palindrome(self):
        if self.head is None:
            return "list is empty"
        slow=self.head
        fast=self.head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        
        if fast:
            curr=slow.next
            slow.next=None
        else:
            curr=slow
        prev=None
        while curr:
            nxt=curr.next
            curr.next=prev
            prev=curr
            curr=nxt
        
        t=prev
        s=self.head
        while t and s:
            if t.data!=s.data:
                return False
            else:
                t=t.next
                s=s.next
        return True
sll=Singlylinked()
a=[1,2]
for i in a:
    sll.insert_at_begining(i)
sll.print_forword()
print()
print(sll.palindrome())
        
```

```python
def kdistinct(data,k):
    return atmost(data,k)-atmost(data,k-1)

def atmost(data,k):
    left=0
    count=0
    hash={}
    
    for right in range(len(data)):
        hash[data[right]]=hash.get(data[right],0)+1
        
        while len(hash)>k:
            hash[data[left]]-=1
            if hash[data[left]]==0:
                del hash[data[left]]
            left+=1
        count+=right-left+1
    return count
a=[1,2,1,2,5]
print(kdistinct(a,2))
```

```python
def minimumlength(nums,target):
    s=0
    left=0
    long=float("inf")
    for i in range(len(nums)):
        s+=nums[i]
        
        while s>=target:
            long=min(i-left+1,long)
            s-=nums[left]
            left+=1
    return -1 if long==float("inf") else long
a=[2,-1,2,3,-5,4]
print(minimumlength(a,4))
```
