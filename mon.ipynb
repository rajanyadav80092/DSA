```python
#question 1

def Earliest(nums):
    x=set()
    for i in nums:
        if i in x:
            return i
        x.add(i)
    return "all unique"
a=[2,1,3,5,3,2]
print(Earliest(a))
#time space O(n) 

def RemoveDup(nums):
    k=0
    for i in nums:
        if i==k:
            k+=1
    return k
nums = [0,0,1,1,1,2,2,3,3,4]
print(RemoveDup(nums))
#time complexity O(n) space O(1) Aisa isliye kiya kyuki data sorted hai to 0 se start hai bhi 

def longestSubstring(nums):
    long=0
    hash={}
    left=0
    for right in range(len(nums)):
        hash[nums[right]]=hash.get(nums[right],0)+1
        while hash[nums[right]]>1:
            hash[nums[left]]-=1
            if hash[nums[left]]==0:
                del hash[nums[left]]
            left+=1
        long=max(right-left+1,long)
    return long
a="abcabcbb"
print(longestSubstring(a))
#time O(n) and space worst case me O(n) warna O(unique) pattern sliding +hashmap

import heapq
def longestk(nums,k):
    heap=[]
    for i in nums:
        heapq.heappush(heap,i)
        
        if len(heap)>k:
            heapq.heappop(heap)
    return heap[0]
nums = [3,2,1,5,6,4]
k = 2
print(longestk(nums,k))
#time complexity O(nlogk) space O(k) pattern kth largest hai to heap yaha minheap hai agar smallest puchha hota to maxheap lagta i jo hai -i se push hota  

def invert(self):
    return self._invert(self.root)
    
def _invert(self,root):
    if root is None:
        return None
    root.left,root.right=root.right,root.left
    self._invert(root.left)
    self._invert(root.right)
    return root
#time complexity O(depth) space O(logdepth) pattern DFS 

def lca(self,p,q):
    return self._lca(self.root,p,q)
    
def _lca(self,root,p,q):
    if not root:
        return None
    if root==p or root==q:
        return root
    left=self._lca(root.left,p,q)
    right=self._lca(root.right,p,q)
        
    if left and right:
        return root
    return left if left else right
#time complexity O(depth) space O(logdepth) pattern DFS 

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
    
class Linked:
    def __init__(self):
        self.head=None
    
    def insert_at_begining(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node
        return
    
    def _merge(self):
        return self.merge(self.head)
    
    def merge(self,head):
        if head is None or head.next is None:
            return head
        slow=head
        fast=head.next
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        mid=slow.next
        slow.next=None
        left=self.merge(head)
        right=self.merge(mid)
        return self.merge_(left,right)
    
    def merge_(self,l1,l2):
        dummy=Node(0)
        temp=dummy
        while l1 and l2:
            if l1.data<l2.data:
                temp.next=l1
                l1=l1.next
            else:
                temp.next=l2
                l2=l2.next
            temp=temp.next
        
        if l1:
            temp.next=l1
        if l2:
            temp.next=l2
        self.head=dummy.next
        return dummy.next
    
    def print_for(self):
        if self.head is None:
            return None
        curr=self.head
        while curr:
            print(curr.data,end=" ")
            curr=curr.next
        print("END")
ll=Linked()
l1=[4,2,1]
l2=[4,3,1]
for i in l1:
    ll.insert_at_begining(i)
for j in l2:
    ll.insert_at_begining(j)
ll.print_for()
ll._merge()
ll.print_for()
#time complexity O(n**2) space O(1) pattern merge sort
```

```python
def nextgreater(nums):
    stack=[]
    rev=[-1]*len(nums)
    for i in range(len(nums)-1,-1,-1):
        while stack and nums[stack[-1]]<nums[i]:
            stack.pop()
        if stack:
            rev[i]=nums[stack[-1]]
        stack.append(i)
    return rev
a=[4,2,5,10]
print(nextgreater(a))
    
```

```python
import random
class Randomized:
    def __init__(self):
        self.arr=[]
        self.pos={}
    
    def insert(self,val):
        if val in self.pos:
            return False
        self.pos[val]=len(self.arr)
        self.arr.append(val)
        return True
    
    def remove(self,val):
        if val not in self.pos:
            return False
        idx=self.pos[val]
        last=self.arr[-1]
        
        self.arr[idx]=last
        self.pos[last]=idx
        
        self.arr.pop()
        del self.pos[val]
    
    def random(self):
        return random(self.arr)
    
    def print_all(self):
        print(self.arr)
        print(self.pos)
ran=Randomized()
ran.insert(1)
ran.insert(2)
ran.insert(3)
ran.insert(4)
ran.remove(2)
ran.print_all()
        
```
